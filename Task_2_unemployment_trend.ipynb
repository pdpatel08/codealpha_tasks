```python
# -------------------------------------------------------------------------
# UNEMPLOYMENT FORECASTING USING ARIMA, LINEAR REGRESSION, AND LSTM
# -------------------------------------------------------------------------

# Install pmdarima for auto_arima functionality
!pip install pmdarima

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------------------
# STEP 1: Load and Prepare Dataset
# ---------------------------------------------------------------
df = pd.read_csv("/content/Unemployment_Rate_upto_11_2020.csv")

# DEBUG: Print DataFrame columns to identify the correct date column name
print("DataFrame columns:", df.columns)
print("DataFrame head:\n", df.head())

df['Date'] = pd.to_datetime(df[' Date'])
df = df.sort_values('Date')
df = df[['Date', ' Estimated Unemployment Rate (%)']]
df.set_index('Date', inplace=True)
df.rename(columns={' Estimated Unemployment Rate (%)': 'Unemployment_Rate'}, inplace=True)

print(df.head())

# ---------------------------------------------------------------
# STEP 2: Visualize Data
# ---------------------------------------------------------------
plt.figure(figsize=(12,5))
plt.plot(df.index, df['Unemployment_Rate'])
plt.title("Unemployment Rate Over Time")
plt.ylabel("Rate (%)")
plt.grid(True)
plt.show()

# ---------------------------------------------------------------
# STEP 3: Train-Test Split (Last 12 months for testing)
# ---------------------------------------------------------------
train = df.iloc[:-12]
test = df.iloc[-12:]

# ===============================================================
# PART A  ARIMA FORECASTING
# ===============================================================

from pmdarima import auto_arima

# Auto ARIMA model fitting
arima_model = auto_arima(train['Unemployment_Rate'], seasonal=False, trace=True)
print(arima_model.summary())

# Predict next 12 months
forecast_arima = arima_model.predict(n_periods=12)
forecast_arima = pd.Series(forecast_arima, index=test.index)

# Plot ARIMA Results
plt.figure(figsize=(12,5))
plt.plot(train.index, train['Unemployment_Rate'], label="Train")
plt.plot(test.index, test['Unemployment_Rate'], label="Actual")
plt.plot(test.index, forecast_arima, label="ARIMA Forecast")
plt.legend()
plt.title("ARIMA Forecast for Unemployment Rate")
plt.show()

# ===============================================================
# PART B  LINEAR REGRESSION FORECASTING
# ===============================================================

from sklearn.linear_model import LinearRegression

df_lr = df.copy()
df_lr['Time_Index'] = np.arange(len(df_lr))

train_lr = df_lr.iloc[:-12]
test_lr = df_lr.iloc[-12:]

model_lr = LinearRegression()
model_lr.fit(train_lr[['Time_Index']], train_lr['Unemployment_Rate'])

pred_lr = model_lr.predict(test_lr[['Time_Index']])

# Plot LR Forecast
plt.figure(figsize=(12,5))
plt.plot(train_lr.index, train_lr['Unemployment_Rate'], label="Train")
plt.plot(test_lr.index, test_lr['Unemployment_Rate'], label="Actual")
plt.plot(test_lr.index, pred_lr, label="LR Forecast")
plt.legend()
plt.title("Linear Regression Forecast for Unemployment Rate")
plt.show()

# ===============================================================
# PART C  LSTM DEEP LEARNING FORECASTING
# ===============================================================

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['Unemployment_Rate']])

# Create sequences (12 months input  next month output)
X, y = [], []
time_steps = 12

for i in range(len(scaled) - time_steps):
    X.append(scaled[i:i+time_steps])
    y.append(scaled[i+time_steps])

X = np.array(X)
y = np.array(y)

# Train-test split
X_train, y_train = X[:-12], y[:-12]
X_test, y_test = X[-12:], y[-12:]

# Build LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

model_lstm = Sequential()
model_lstm.add(LSTM(64, activation='relu', input_shape=(12,1)))
model_lstm.add(Dense(1))
model_lstm.compile(loss='mse', optimizer='adam')
model_lstm.summary()

# Train LSTM
model_lstm.fit(X_train, y_train, epochs=50, batch_size=8, verbose=1)

# Predict using LSTM
y_pred_lstm = model_lstm.predict(X_test)
y_pred_lstm = scaler.inverse_transform(y_pred_lstm)
y_test_real = scaler.inverse_transform(y_test)

# Plot LSTM Forecast
plt.figure(figsize=(12,5))
plt.plot(test.index, y_test_real, label="Actual")
plt.plot(test.index, y_pred_lstm, label="LSTM Forecast")
plt.legend()
plt.title("LSTM Forecast for Unemployment Rate")
plt.show()

# -------------------------------------------------------------------------
# END OF FULL PROJECT CODE
```
